```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

```

```python
data = np.loadtxt('data.txt')
movie_data = []
with open('movies.txt', encoding="ISO-8859-1") as movie_file:
    for line in movie_file:
            line = line.split('\t')
            movie_data.append(line)
print(movie_data)
movie_data = np.array(movie_data)
ratings = np.array(data[:, 2])
sns.distplot(ratings, kde=0)

# for line in i

#plt.hist(ratings)

```

```python
movie_indices = np.array(data[:, 1])

movie_counts = {}

for elem in movie_indices:
    if elem not in movie_counts:
        movie_counts[elem] = 1
    else:
        movie_counts[elem] += 1

```

```python
averages = {}

for i in range(len(movie_indices)):
    elem = movie_indices[i]
    if elem not in averages:
        averages[elem] = ratings[i]
    else:
        averages[elem] += ratings[i]

for elem in averages:
    averages[elem] /= movie_counts[elem]


```

```python
top_ten_averages = sorted(averages, key=lambda x: averages[x])[::-1][:10]

```
